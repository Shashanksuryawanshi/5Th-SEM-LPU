# Quartiles: The 25th (Q1) and 75th (Q3) percentiles, showing the spread of the middle 50% of the data.

# Whiskers: Extend to show the range of the data, typically 1.5 times the interquartile range (IQR) from the quartiles.

### Filling the missing values  - Imputation (NUMERICAL DATA)

```python
df3 = df.copy()
```

```python
df3.isna().sum()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3['reviews_per_month'] = df3['reviews_per_month'].fillna(mean_value)

```

```python
df3.head()
#In this updated dataset you can check That reviews column is filling with a mean value
```

```python
most_frequent = df3["last_review"].value_counts().index[0]
```

```python
df3["last_review"] = df3["last_review"].fillna(most_frequent)
```
